```python
#importações

import pandas as pd
import numpy as np
import random
from datetime import datetime
```

```python
# ID do abastecimento
#Serão 10000 linhas de dados

num_id=np.arange(1,10001,1,int)
```

```python
# Volume
#Será valores aleatórios de 1 a 50, representando a quantidade de litros abastecida

volume=np.zeros(10000)
for i in range(10000):
    volume[i]=round((random.random()*50),0)
    
```

```python
#Combustivel
#Poderá ser gasolina, etanol e gasolina aditivada, serão sorteadas uma dessas para cada uma das linhas
s=['etanol', 'gasolina', 'gasolina_aditivada']
comb=[]

for i in range(10000):
    comb.append(random.choice(s))
    

```

```python
#Data e hora da compra
#Será sorteado os parâmetro de data e hora

datas=[]

for i in range(10000):

    #lista com o número de todos os meses de um ano
    l_mes=np.arange(1,13,1,int)
    #escolhendo um mês
    mes=random.choice(l_mes)
    #lista com o número de todos os possiveis dias de um mes
    l_dias=np.arange(1,31,1,int)
    d_30=[4,6,9,11]

    #escolhendo o dia conforme o mes
    if mes==2:
        dia=random.choice(l_dias[:28])
    elif mes in d_30:
        dia=random.choice(l_dias[:30])
    else:
        dia=random.choice(l_dias[:31])

    #serão considerado os anos de 2020, 2021, 2022 
    #escolhendo um ano
    ano=random.choice(np.arange(2020,2023,1,int))


    #sorteando uma hora e um minuto
    hora=random.choice(np.arange(0,24,1,int))
    minutos=random.choice(np.arange(0,60,1,int))


    data=str(ano)+'/'+str(mes)+"/"+str(dia)+" "+str(hora)+":"+str(minutos)
    data_final=datetime.strptime(data,'%Y/%m/%d %H:%M')

    datas.append(data_final.strftime('%Y/%m/%d %H:%M'))
    
```

```python
#Criando o dataframe

df1=pd.DataFrame(num_id,columns=['id_abas'])
df1['data_hora']=datas
df1['volume']=volume
df1['combustivel']=comb
df1.head()
```

```python
#Valor

#O valor será calculado conforme o tipo de combustivel e o volume abastecido

l_valor=[]
#preços: gasolina:5,00 etanol 3,50 gasolina aditivada 5,5
for i in range(len(df1)):
    if df1['combustivel'][i]=="gasolina_aditivada":
        l_valor.append((df1['volume'][i])*5.5)
    elif(df1['combustivel'][i]=="gasolina"):
        l_valor.append((df1['volume'][i])*5)
    else:
        l_valor.append((df1['volume'][i])*3.5)
        
df1["valor"]=l_valor

df1.head(10)
```

```python
df1.to_csv("dados_abastecimento_projeto",index=False)
```
